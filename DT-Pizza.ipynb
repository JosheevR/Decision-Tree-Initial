```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt

from sklearn.tree import plot_tree
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

```python
data = pd.read_csv("Pizza-Price.csv")
data.replace({'yes': True, 'no': False}, inplace=True)
print(data)
```

```python
y = data['Price']
X = data.drop(columns=['Price', 'Restaurant'])
```

```python
y_train, y_test = y.iloc[:-2], y.iloc[-2:]
X_train, X_test = X.iloc[:-2], X.iloc[-2:]
```

```python
decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
```

```python
# model = XGBRegressor()
# model.fit(X_train, y_train)
# y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Regression Mean Squared Error:", mse)
```

```python
plt.figure(figsize=(20,10))
plot_tree(decision_tree, feature_names=X.columns, filled=True, rounded=True)
plt.show()
```
